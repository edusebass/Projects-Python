```python
a = open("constitucion.txt")
b = a.read()

puntuacion = '''!()-[]{};:'"\,<>./?@#$%^&*_~1234567890'''
b_sin_puntuacion = ""

for caracter in b:
    if caracter not in puntuacion:
        b_sin_puntuacion += caracter

palabras = b_sin_puntuacion.lower().split()

frecuencia = {}

for palabra in palabras:
    if palabra != "":
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1

palabras_ordenadas = sorted(frecuencia.items())

for palabra, repeticiones in palabras_ordenadas:
    print(f'({palabra}, {repeticiones})')

a.close()
```
