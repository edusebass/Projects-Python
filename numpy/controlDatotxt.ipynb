```python
def creartxt():
    archi = open('datos.txt', 'w')
    archi.close()
    
def grabartxt():
    archi = open('datos.txt', 'a')
    archi = open('datos.txt','a') 
    archi.write('Linea 1\n') 
    archi.write('Linea 2\n') 
    archi.write('Linea 3\n') 
    archi.close()
    
creartxt()
grabartxt()    
```

```python
def leertxt(): 
    archi = open('datos.txt','r') 
    linea = archi.readline() 
    while linea != "":
        print(linea)
        linea = archi.readline() 
    archi.close()
    
def leertxtenlista(): 
    archi = open('datos.txt','r') 
    lineas = archi.readlines() 
    print (lineas)
    archi.close()
    
leertxt()
```

```python
#ejemplo
cadenaPalabras = "Holal quetal hola que mas que pasa mas esto es importante, hola"
```

```python
listaPalabras = cadenaPalabras.split()
print(listaPalabras)
```

```python
frecuenciaPalab = []
```

```python
for w in listaPalabras:
    frecuenciaPalab.append(listaPalabras.count(w))
```

```python

a = open("datac.txt")
b = a.read()
palabras = b.split()

frecuencia = {}

for palabra in palabras:
    palabra = palabra.strip()
    if palabra != "":
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
    
palabras_ordenadas = sorted(frecuencia.items())

for palabra, repeticiones in frecuencia.items():
    print(palabra, repeticiones)
```
